### Remove the punctuation

```python
def remove_punctuation(text):
    if isinstance(text, str):
        return text.translate(str.maketrans('', '', string.punctuation)).replace('\n', ' ')
    else:
        return text  
df['Review'] = df['Review'].apply(remove_punctuation)
```

### Importing the file slang.txt that can be used to decode the slang language

```python
with open('slang.txt', 'r') as file:
    lines = file.readlines()
```

```python
slangdict = {}
for line in lines:
    if '=' in line:
        key, value = line.strip().split('=')
        slangdict[key] = value
```

```python
slangdict
```

```python
def replace_slang(text):
    # Check if the text is not NaN
    if isinstance(text, str):
        for slang, expansion in slangdict.items():
            text = text.replace(slang, expansion)
    return text
df['Review'] = df['Review'].apply(replace_slang)
```

### Remove Emojis

```python
import re

def remove_emojis(text):
    if isinstance(text, str):
        emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  
                               u"\U0001F300-\U0001F5FF" 
                               u"\U0001F680-\U0001F6FF" 
                               u"\U0001F1E0-\U0001F1FF"
                               u"\U00002500-\U00002BEF" 
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)
    else:
        return text
df['Review'] = df['Review'].apply(remove_emojis)
```

### Lemmatization

```python
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk
nltk.download('wordnet')

def lemmatize_tokens(tokens):
    if isinstance(tokens, list):
        lemmatizer = WordNetLemmatizer()
        lemmatized_tokens = [lemmatizer.lemmatize(token, pos=wordnet.VERB) for token in tokens]
        return lemmatized_tokens
    else:
        return tokens
df['Review'] = df['Review'].apply(lemmatize_tokens)
```
